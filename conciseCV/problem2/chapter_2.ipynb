```python
import numpy as np
import cv2
import matplotlib.pyplot as plt 
```

```python
def resize(img, scale=0.1):
    w = round(img.shape[1] * scale)
    h = round(img.shape[0] * scale)
    new_img = cv2.resize(img, (w,h))
    return new_img

def scale(img):
    # scale image value to the range [0, 256]
    min_val = np.amin(img)
    new_img = img - min_val
    max_val = np.amax(new_img)
    new_img = np.rint((new_img / max_val) * 256)
    return new_img.astype("uint8")
    
    
```

```python
img_path = "/home/hieppm/vmShare/greyImg/p0.jpg"
img = cv2.imread(img_path, 0)
```

```python
img2 = resize(img, 0.4)
```

```python
# Add noise
noise = np.ones([img2.shape[0], img2.shape[1]])
_ = cv2.randn(noise, 0, 1)
c = 20
noised_img2 = np.rint(img2 + c * noise)
noised_img2 = scale(noised_img2)
```

```python
hist = cv2.calcHist([noised_img2], [0], None, [256], [0, 255])
plt.plot(hist)
plt.show()
```
